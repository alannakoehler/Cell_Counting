```python
# Import necessary programs
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Try to import data from .csv files generated by ImageJ macro
```

```python
# Define important variables at the top
layer = list(range(1,11)) # Generate a list of values 1 to 10
brdu = 'e11.5'
surv = 'p7.5'
slide = '67A'
red_file = "/Users/akoehler/Desktop/Cell_Counting/Results/Red_67A.csv"
yellow_file = "/Users/akoehler/Desktop/Cell_Counting/Results/Yellow_67A.csv"
green_file = "/Users/akoehler/Desktop/Cell_Counting/Results/Green_67A.csv"


# Import green file
red = pd.read_csv(red_file)
yellow = pd.read_csv(yellow_file)
green = pd.read_csv(green_file)
red
```

```python
# Insert BrdU, Survival, and Section columns
red.insert(1, 'BrdU', brdu)
red.insert(2, 'Survival', surv)
red.insert(3, 'Section', slide)
red.insert(3, 'Layer', layer)
# Delete unnecessary information
del red['Slice']
red.drop(['Total Area', 'Average Size', '%Area', 'Mean'], axis = 1, inplace = True)
# Rename column for color indication
red.rename(columns = {'Count': 'Count (Red)'}, inplace = True)
red
```

```python
# Preview yellow .csv
yellow
```

```python
# Add yellow counts to dataframe
yellow = yellow['Count']
yellow
```

```python
red_yellow = pd.concat([red, yellow], axis = 1)
red_yellow.rename(columns = {'Count': 'Count (Yellow)'}, inplace = True)
red_yellow
```

```python
# Preview green .csv
green
```

```python
# Add green counts to dataframe
green = green['Count']
green
```

```python
red_yellow_green = pd.concat([red_yellow, green], axis = 1)
red_yellow_green.rename(columns = {'Count': 'Count (Green)'}, inplace = True)
red_yellow_green 
```

```python
# Define a path and filename that updates with slide number
new_path = '/Users/akoehler/Desktop/Cell_Counting/Results/Per_Slide_CSVs/red_yellow_green' + slide +'.csv'
new_path
```

```python
# Export to .csv file
export_csv = red_yellow_green.to_csv (new_path, index = None, header=True)
```
